# Problema #1: Calculadora de propina inteligente
# Problema
Desarrollar un programa que solicite al usuario el total de una cuenta y el porcentaje de propina que desea dejar. El programa debe calcular el valor de la propina y el monto total a pagar.

# Entrada

Total de la cuenta (float).
Porcentaje de propina (float).

# Proceso

1-Solicitar el total de la cuenta.

2-Solicitar el porcentaje de propina.

3-Convertir los datos a float.

4-Calcular la propina.

5-Calcular el total final.

6-Mostrar los resultados usando f-strings.

7-Validar entradas con try-except.

# Salida esperada

Mostrar:
Total de la cuenta.
Valor de la propina.
Total a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 100,15| propina = 15, total = 115| |
| Caso límite | 0, 10| propina = 0, total = o| |
| Caso inválido |abc, 15 |msj de error| |


```python
try:
    total = float(input("Ingrese el total de la cuenta: "))
    porcentaje = float(input("Ingrese el porcentaje de propina: "))

    propina = total * (porcentaje / 100)
    total_pagar = total + propina

    print(f"\nTotal de la cuenta: ${total:.2f}")
    print(f"Propina ({porcentaje}%): ${propina:.2f}")
    print(f"Total a pagar: ${total_pagar:.2f}")

except ValueError:
    print("Error: Debe ingresar valores numéricos válidos.")
```

# Problema #2: Detector de número par/impar

# Problema

Crear un programa que solicite un número entero e indique si es par o impar.

# Entrada

Número entero (int).

# Proceso

1-Solicitar un número entero.

2-Convertirlo a tipo int.

3-Calcular el residuo de dividir entre 2 usando %.

4-Si el residuo es 0, el número es par.

5-De lo contrario, es impar.

6-Validar entradas con try-except.

# Salida esperada

Mostrar un mensaje indicando si el número es par o impar.

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 8|es par| |
| Caso límite | 0| es par| |
| Caso inválido | hola| msj de error| |


```python
try:
    numero = int(input("Ingrese un número entero: "))

    if numero % 2 == 0:
        print(f"El número {numero} es par.")
    else:
        print(f"El número {numero} es impar.")

except ValueError:
    print("Error: Debe ingresar un número entero válido.")
```

# Problema #3: Clasificador de edad

Problema
Desarrollar un programa que solicite la edad de una persona y la clasifique como menor de edad, adulto o adulto mayor.

# Entrada

Edad (int).

# Proceso

Solicitar la edad.

Convertir el dato a entero.

Validar que la edad no sea negativa.

Aplicar las siguientes reglas:

Menor de edad: menos de 18 años.

Adulto: entre 18 y 64 años.

Adulto mayor: 65 años o más.

Mostrar el resultado.

Manejar errores con try-except.

# Salida esperada

Mostrar la clasificación correspondiente.

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | 30| adulto| |
| Caso límite | 65| adulto mayor| |
| Caso inválido | -5| msj de error| |


```python
try:
    edad = int(input("Ingrese su edad: "))

    if edad < 0:
        print("Error: La edad no puede ser negativa.")
    elif edad < 18:
        print("La persona es menor de edad.")
    elif edad < 65:
        print("La persona es adulta.")
    else:
        print("La persona es adulto mayor.")

except ValueError:
    print("Error: Debe ingresar una edad válida.")
```

# Problema #4: Validador de compra con descuento

# Problema

Crear un programa que solicite el nombre de un cliente, el monto de una compra y si es cliente frecuente. Dependiendo de las condiciones, se aplicará un descuento.

# Reglas

Si la compra es mayor o igual a $100 y el cliente es frecuente → 15% de descuento.

Si solo cumple una de las condiciones → 5% de descuento.

Si no cumple ninguna → 0% de descuento.

# Entrada

Nombre del cliente (string).

Monto de compra (float).

Cliente frecuente (sí/no).

# Proceso

Solicitar nombre.

Solicitar monto de compra.

Solicitar si es cliente frecuente.

Limpiar el texto con .strip().lower().

Evaluar condiciones usando and y or.

Calcular descuento.

Mostrar resultados mediante f-strings.

Validar el monto con try-except.

# Salida esperada

Mostrar:

Nombre del cliente.

Porcentaje de descuento.

Valor descontado.

Total a pagar.

## Pruebas manuales

| Escenario | Datos de entrada | Resultado esperado | Resultado obtenido |
|-----------|------------------|--------------------|--------------------|
| Caso normal | juan, 200, si| 15% de descuento| |
| Caso límite | ana, 100, no| 5% de descuento| |
| Caso inválido | pedro, abc, si| msj de error| |


```python
try:
    nombre = input("Ingrese el nombre del cliente: ").strip()

    monto = float(input("Ingrese el monto de la compra: "))

    frecuente = input(
        "¿Es cliente frecuente? (si/no): "
    ).strip().lower()

    if monto >= 100 and frecuente == "si":
        descuento = 0.15

    elif monto >= 100 or frecuente == "si":
        descuento = 0.05

    else:
        descuento = 0

    valor_descuento = monto * descuento
    total_pagar = monto - valor_descuento

    print(f"\nCliente: {nombre}")
    print(f"Descuento aplicado: {descuento*100:.0f}%")
    print(f"Valor descuento: ${valor_descuento:.2f}")
    print(f"Total a pagar: ${total_pagar:.2f}")

except ValueError:
    print("Error: El monto debe ser un número válido.")
```
